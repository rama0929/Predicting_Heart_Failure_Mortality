```python
#!/usr/bin/env python

# make sure to install these packages before running:
# pip install pandas
# pip install sodapy

import pandas as pd
from sodapy import Socrata

# Unauthenticated client only works with public data sets. Note 'None'
# in place of application token, and no username or password:
client = Socrata("chronicdata.cdc.gov", None)

# Example authenticated client (needed for non-public datasets):
#client = Socrata("chronicdata.cdc.gov",
                #"LQGVHy6aGXfJugJhhgTPhOdEq",
                #userame="T.Ormanoski@gmail.com",
                #password="Zoey1446!")

# First 2000 results, returned as JSON from API / converted to Python list of
# dictionaries by sodapy.
results = client.get("i2vk-mgdh", limit=100000)

# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
# Drop "datasource" column
results_df = results_df.drop(columns = "datasource")
results_df.head(10)
```

```python
# Drop "geographiclevel" column
results_df = results_df.drop(columns = "geographiclevel")
results_df.shape
```

```python
# Drop "stratificationcategory1" columns
results_df = results_df.drop(columns = "stratificationcategory1")

# Rename the "stratification1" column "Gender"
results_df = results_df.rename(columns = {'stratification1':'Gender'})

results_df.head(10)
```

```python
# Drop "stratificationcategory2" columns
results_df = results_df.drop(columns = "stratificationcategory2")

# Rename the "stratification1" column "Gender"
results_df = results_df.rename(columns = {'stratification2':'Race/Gender'})

results_df.head(10)
```

```python
# Drop last four columns
results_df = results_df.drop(results_df.iloc[:, 13:17], axis = 1)
results_df.head()
```

```python
# Drop "topicid" and "locationid" columns

results_df = results_df.drop(results_df.iloc[:, 10:12], axis = 1)
results_df.head()
```

```python
# Get shape before dropping null values
results_df.shape
```

```python
# Drop null values
results_df = results_df.dropna()

# Get shape after
results_df.shape
```
