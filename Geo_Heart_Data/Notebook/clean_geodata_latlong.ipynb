```python
import pandas as pd
```

```python
#Import the GeoData file
geodata_file = f'C://Users//abaez//Desktop//Predicting_Heart_Failure_Mortality//Geo_Heart_Data//Notebook//GeoData.csv'
```

```python
#Convert data to a dataframe
geodata_df = pd.read_csv(geodata_file, low_memory=False)
```

```python
#Check the dataframe
geodata_df.head()
```

```python
#Rename the 'locationabbr' column to 'state'
geodata_df.rename(columns = {'locationabbr':'state'}, inplace = True)
```

```python
#Check the column name
geodata_df.head()
```

```python
#Check the unique values in the column 'locationdesc' to see if there is a more specific title for the column
geodata_df.locationdesc.unique().tolist()
```

```python
#Rename the column 'locationdesc' to 'city'
geodata_df.rename(columns = {'locationdesc':'city'}, inplace = True)
```

```python
#Check the new column name
geodata_df.head()
```

```python
#Convert the last column into a series
latlong_series = geodata_df['location_1'].map(eval)
latlong_series
```

```python
#Turn latlong_series into a dataframe
latlong_df = pd.DataFrame.from_dict(latlong_series)
latlong_df
```

```python
#Separate the key, value pairs in the dictionary into two columns with the key as the column name
lat_long_df = pd.json_normalize(latlong_df['location_1'])
lat_long_df
```

```python
#Combine the two dataframes into one
new_geodata_df = pd.merge(geodata_df, lat_long_df,left_index = True, right_index = True)
new_geodata_df
```

```python
#Save new_geodata_df to a .csv
new_geodata_df.to_csv('C://Users//abaez//Desktop/Predicting_Heart_Failure_Mortality/Geo_Heart_Data/Notebook/clean_geodata_latlong.csv', index=False)
```
